```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# cap = cv2.VideoCapture('./video/094448.mp4')
cap = cv2.VideoCapture('./video/093557.mp4')
count = 0
count_list = []
pixel_value_mean_list = []
```

```python
while(cap.isOpened()):
    end_flag, frame = cap.read()
#     print (frame)

    if frame is None:
        break;

    # 回転部分の切り取り
    pixel_value = frame[289, 438]
    pixel_value_mean = np.average(pixel_value)
    
#     print (pixel_value,':', pixel_value_mean)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    count += 1
    count_list.append(count)
    pixel_value_mean_list.append(pixel_value_mean)
    
# print (count_list)
# print (pixel_value_mean_list)
    
cap.release()
cv2.destroyAllWindows()
```

```python
x = count_list[3500:]
y = pixel_value_mean_list[3500:]
plt.figure(dpi=200)
plt.title('velocity')
plt.plot(x, y)
# plt.savefig('./test.png')
```
