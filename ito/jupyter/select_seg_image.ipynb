```python
def progress_time(time):
    hour , minu ,sec  = time.split(':')
    sec = int(sec) + 1
    minu = int(minu)
    hour = int(hour)
    if sec == 60 :
        sec = 0
        minu += 1
        if minu >= 60:
            minu = 0
            hour += 1

    return str(hour) + ':' + str(minu) +':' + str(sec)
```
